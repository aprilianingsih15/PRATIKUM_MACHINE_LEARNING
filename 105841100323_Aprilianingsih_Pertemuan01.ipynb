```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.datasets import load_wine

# =================================================================
# NOMOR 1: Statistik Array Random
# =================================================================
print("--- NOMOR 1: Statistik Array Random ---")
# Membuat array berisi 20 angka acak antara 0 sampai 100
np.random.seed(42)
array_random = np.random.randint(0, 101, 20)

# Menghitung Mean, Median, dan Standar Deviasi
mean_val = np.mean(array_random)
median_val = np.median(array_random)
std_val = np.std(array_random)

print(f"Array Random (20 angka): \n{array_random}")
print(f"Hasil Perhitungan:")
print(f"- Mean (Rata-rata)  : {mean_val:.2f}")
print(f"- Median (Nilai Tengah): {median_val}")
print(f"- Std Dev (Simpangan Baku): {std_val:.2f}\n")
```

Pada bagian ini saya membuat array berisi 20 angka acak antara 0 sampai 100, kemudian menghitung mean, median, dan standar deviasi untuk mengetahui karakteristik data tersebut.
Dari hasil perhitugan diperoleh nilai rata - rata, median, dan standar deviasi. Mean menunjukkan nilai rata- rata keseluruhan data, median menunjukkan nilai tengah, dan standar deviasi menunjukkan tingkat penyebaran data.

```python
# =================================================================
# NOMOR 2: DataFrame Nilai Mahasiswa & 3 Nilai Tertinggi
# =================================================================
print("--- NOMOR 2: DataFrame Nilai Mahasiswa ---")
# Membuat data 10 mahasiswa dengan nama yang diminta
data_nilai = {
    'Mahasiswa': ['Amanda', 'Hizra', 'Fahira', 'Indah', 'Aprilia', 'Hasriana', 'Dina', 'Ruhi', 'Ivan', 'Joko'],
    'Nilai': [85, 92, 78, 88, 95, 70, 82, 90, 89, 75]
}
df_mhs = pd.DataFrame(data_nilai)

# Mencari 3 nilai tertinggi
top_3 = df_mhs.nlargest(3, 'Nilai')

print("Tabel Nilai Mahasiswa:")
print(df_mhs)
print("\n3 Mahasiswa dengan Nilai Tertinggi:")
print(top_3, "\n")
```

Pada bagian ini saya membuat DataFrame berisi 10 mahasiswa dan nilainya. Kemudian saya mencari 3 mahasiswa dengan nilai tertinggi menggunakan fungsi nlargest().
Dari hasil yang diperoleh, terlihat tiga mahasiswa dengang nilai tertinggi berdasarkan data yang ada. Fungsi nlargest() mempermudah pencarian nilai terbesar tanpa perlu mengurutkan seluruh data secara manual.

```python
# =================================================================
# NOMOR 3: Visualisasi Data (Grafik Batang & Grafik Bulat)
# =================================================================
print("--- NOMOR 3: Visualisasi Data ---")
plt.figure(figsize=(15, 7))

# --- Grafik A: Bar Chart (Grafik Batang) ---
plt.subplot(1, 2, 1)
df_sorted = df_mhs.sort_values('Nilai', ascending=False)
bars = plt.bar(df_sorted['Mahasiswa'], df_sorted['Nilai'], color='skyblue', edgecolor='navy')

# Menambahkan angka nilai di atas setiap batang
for bar in bars:
    yval = bar.get_height()
    plt.text(bar.get_x() + bar.get_width()/2, yval + 0.5, yval, ha='center', va='bottom', fontsize=10)

plt.title('Perbandingan Nilai Mahasiswa (Bar Chart)', fontsize=14)
plt.xlabel('Nama Mahasiswa')
plt.ylabel('Nilai')
plt.xticks(rotation=45)

# --- Grafik B: Pie Chart (Grafik Bulat) ---
plt.subplot(1, 2, 2)
# Memberikan efek 'explode' (terpotong keluar) untuk nilai tertinggi
explode = [0.1 if x == df_mhs['Nilai'].max() else 0 for x in df_mhs['Nilai']]

plt.pie(df_mhs['Nilai'], labels=df_mhs['Mahasiswa'], autopct='%1.1f%%', startangle=140,
        colors=plt.cm.Paired.colors, explode=explode, shadow=True)

plt.title('Proporsi Nilai Mahasiswa (Pie Chart)', fontsize=14)


plt.tight_layout()
plt.show() # Tampilkan grafik
print("Grafik berhasil ditampilkan.\n")
```

Pada bagian ini dilakukan visualisasi data menggunakan grafik batang untuk membandingkan nilai mahasiswa dan grafik lingkaran untuk melihat proporsi nilai masing - masing mahasiswa.
Dari hasilnya, grafik batang menunjukkan perbandingan nilai antara mahasiswa secara jelas. Sedangkan grafik lingkaran menunjukkan presentase kontrobusi masing - masing nilai terhadap keseluruhan data. Mahasiswa dengan nilai tertinggi terlihat lebih menonjol pada grafik.

```python
# =================================================================
# NOMOR 4: Load Dataset Wine & 3 Insight
# =================================================================
print("--- NOMOR 4: Insight Dataset Wine ---")
wine = load_wine()
df_wine = pd.DataFrame(wine.data, columns=wine.feature_names)
df_wine['target'] = wine.target

# Analisis singkat
avg_alcohol = df_wine['alcohol'].mean()
class_counts = df_wine['target'].value_counts().to_dict()

print(f"Dataset Wine berhasil dimuat ({df_wine.shape[0]} baris).")
print("3 Insight Singkat:")
print(f"1. Rata-rata kadar alkohol pada seluruh sampel adalah {avg_alcohol:.2f}%.")
print(f"2. Terdapat 3 kategori kelas anggur dengan distribusi sampel: {class_counts}.")
print("3. Fitur kimia seperti 'proline' dan 'flavanoids' merupakan indikator utama dalam pemisahan kelas target.")

# Simpan data ke CSV
df_mhs.to_csv('data_mahasiswa_tugas.csv', index=False)
```

Pada bagian ini saya memuat dataset Wine dari sklearn, kemudian mengubahnya menjadi DataFrame untuk dianalisis. selanjutnya dihitung rata - rata kadar alkohol dan distribusi jumlah data pada masing - masing kelas.
Dari hasil analisis dataset Wine, diperoleh rata -rata kadar alkohol serta distribusi data pada tiga kelas anggur. Dataset ini dapat digunakan untuk latihan klasifikasi karena memiliki beberapa fitur kimia yang membedakan tiap kelasnya.
